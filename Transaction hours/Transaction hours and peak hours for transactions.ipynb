```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import squarify 
import seaborn as sb
```

```python
df = pd.read_csv("Fraudulent_E-Commerce_Transaction_Data.csv")
```

```python
fraud_df = df[df['Is Fraudulent'] == 1]
fraud_df.head()
```

```python
amount = fraud_df["Transaction Hour"].value_counts()
df_amount = pd.DataFrame(amount)
df_amount.head()
```

```python
time_fraud = fraud_df.sort_index()
time_fraud.head()
```

```python
df_amount.reset_index(inplace=True)
```

```python
df_amount
```

```python
df_amount["prop"] = round((df_amount['count']/df_amount['count'].sum())*100, 2)
df_amount['label'] = 'hr ' + df_amount['Transaction Hour'].astype(str) + ' \n (' + df_amount['prop'].astype(str) + '%)'
df_amount.head()
```

```python
# plot it
plt.figure(figsize=(10,7))
squarify.plot(sizes=df_amount['count'], label = df_amount['label'], color = sb.color_palette("rocket", len(df_amount['count'])), alpha=.8)
plt.axis('off')
plt.show()
```
