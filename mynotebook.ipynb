```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Fraudulent_E-Commerce_Transaction_Data.csv')
df2 = pd.read_csv('Fraudulent_E-Commerce_Transaction_Data_2.csv')
```

```python
df
```

```python
df2
```

```python
merged_df = pd.concat([df,df2], ignore_index=True)
```

```python
merged_df
```

```python
reg_txns = merged_df[merged_df['Is Fraudulent'] == 0]
fraud_df = merged_df[merged_df['Is Fraudulent'] == 1]
```

```python
time_fraud = fraud_df['Transaction Hour'].value_counts()
time_fraud = time_fraud.sort_index()

# Plotting the bar chart
plt.figure(figsize=(10, 6))
time_fraud.plot(kind='bar', color='skyblue')
plt.title('Fraud Transactions by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Fraud Transactions')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
fraudulent_transactions = merged_df[merged_df['Is Fraudulent'] == 1]
# Count the number of fraudulent transactions for each account age day
fraudulent_transactions_count = fraudulent_transactions.groupby('Account Age Days').size()

# Plotting Fraudulent Transactions vs Account Age Days using Line Plot
plt.figure(figsize=(10, 6))
plt.plot(fraudulent_transactions_count.index, fraudulent_transactions_count.values, color='red', marker='o')
plt.title('Fraudulent Transactions vs Account Age Days')
plt.xlabel('Account Age Days')
plt.ylabel('Number of Fraudulent Transactions')
plt.grid(True)
plt.show()
```

```python
reg_txns_count = reg_txns.shape[0]
fraud_count = fraud_df.shape[0]

# Data to plot
labels = ['Regular Transactions', 'Fraudulent Transactions']
sizes = [reg_txns_count, fraud_count]
colors = ['skyblue', 'salmon']
explode = (0, 0.1)  # Explode the Fraudulent Transactions slice

# Plot
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Regular Transactions vs Fraudulent Transactions')

# Draw a circle at the center of the pie to make it a donut chart
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')  
plt.show()
```
